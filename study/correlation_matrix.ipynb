```python
import os
import xml.etree.ElementTree as ET
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def parse_xml(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    data = []

    for obj in root.findall('object'):
        bndbox = obj.find('bndbox')
        xmin = float(bndbox.find('xmin').text)
        ymin = float(bndbox.find('ymin').text)
        xmax = float(bndbox.find('xmax').text)
        ymax = float(bndbox.find('ymax').text)
        data.append({'xmin': xmin, 'ymin': ymin, 'xmax': xmax, 'ymax': ymax})

    return data

def gather_data_from_directory(directory):
    all_data = []

    for filename in os.listdir(directory):
        if filename.endswith('.xml'):
            file_path = os.path.join(directory, filename)
            data = parse_xml(file_path)
            all_data.extend(data)

    return pd.DataFrame(all_data)

def plot_correlogram(dataframe):
    correlation_matrix = dataframe.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='viridis')
    plt.title('Correlograma das Coordenadas')
    plt.show()

def main():
    directory = 'annotations'  # Substitua pelo caminho do seu diretório
    data = gather_data_from_directory(directory)
    plot_correlogram(data)

if __name__ == '__main__':
    main()
```

```python
import os
import xml.etree.ElementTree as ET
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np

# Diretório contendo os arquivos XML
xml_dir = "f:/x_annotations/"

# Lista para armazenar todos os rótulos
labels = []

# Percorre todos os arquivos no diretório
for filename in os.listdir(xml_dir):
    if filename.endswith(".xml"):
        xml_file = os.path.join(xml_dir, filename)
        
        # Parseia o XML
        tree = ET.parse(xml_file)
        root = tree.getroot()
        
        # Extrai todos os rótulos das tags <name>
        for obj in root.findall('object'):
            label = obj.find('name').text
            labels.append(label)

# Contagem de frequência dos rótulos
label_counts = Counter(labels)

# Criação do histograma
plt.figure(figsize=(10, 6))

# Gera um mapa de cores baseado na paleta viridis
colors = plt.cm.viridis(np.linspace(0, 1, len(label_counts)))

# Plota as barras com a paleta viridis
plt.bar(label_counts.keys(), label_counts.values(), color=colors)
plt.xlabel('Rótulos')
plt.ylabel('Frequência')
plt.title('Histograma de Frequência dos Rótulos')
plt.xticks(rotation=45, ha="right")  # Rotaciona os rótulos para facilitar a leitura
plt.tight_layout()

# Salvar o histograma como imagem (opcional)
plt.savefig("histograma_rotulos_viridis.png")

# Exibe o histograma
plt.show()
```
