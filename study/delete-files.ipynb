```python
import os
import xml.etree.ElementTree as ET
```

```python
# Lista de valores possíveis para a tag <name>
valid_names = [
    "Transverse Cracks", "Block Cracks", "Edge Cracks",
    "Rutting", "Depression", "Corrugation and shoving"
]
```

```python
# Função para verificar e deletar arquivos
def process_annotations():
    for filename in os.listdir(annotations_dir):
        
        if filename.endswith('.xml'):
            xml_path = os.path.join(annotations_dir, filename)
            tree = ET.parse(xml_path)
            root = tree.getroot()
            
            # Verifica se há uma tag <object> com a tag <name> correspondente
            for obj in root.findall('object'):
                name = obj.find('name').text
                if name in valid_names:
                    image_filename = root.find('filename').text
                    image_path = os.path.join(images_dir, image_filename)

                    # Deleta o arquivo de imagem correspondente se existir
                    if os.path.exists(xml_path):
                        os.remove(image_path)
                        os.remove(xml_path)
                        print(f'Deleted: {image_path}')
                        print(f'Deleted: {xml_path}')

                    break
```

```python
# Diretórios
annotations_dir = 'D:/python/tcc1/annotations_o5c'
images_dir = 'D:/python/tcc1/images_o5c'
```

```python
process_annotations()
```
