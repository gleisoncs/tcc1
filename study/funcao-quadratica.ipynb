```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar dados para a função quadrática
x = np.linspace(-3, 3, 10)
print(x)
y = x**2
print(y)

# Criar o gráfico da função quadrática
plt.figure(figsize=(8,6))
plt.plot(x, y, label=r'$f(x) = x^2$', color='blue')

# Adicionar o ponto mínimo (x=0, y=0)
plt.scatter(0, 0, color='red', zorder=5)
plt.text(0, 0.5, r'Mínimo global $(0, 0)$', fontsize=12, color='red', ha='center')

# Configurações do gráfico
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)
plt.title('Gráfico da Função Quadrática $f(x) = x^2$', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.grid(True)
plt.legend()

# Mostrar o gráfico
plt.show()
```
