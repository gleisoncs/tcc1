# HeatMap 

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Dados da tabela
data = {
    'Labels': ['Alligator Crack', 'Block Crack', 'Corrugation', 'Edge Cracks', 'Longitudinal Crack',
               'Manhole', 'Patching', 'Potholes', 'Raveling', 'Transverse Crack'],
    'Precision': [0.36, 0.00, 0.00, 0.00, 0.20, 0.71, 0.20, 0.00, 0.40, 0.22],
    'Recall': [0.78, 0.00, 0.00, 0.00, 0.44, 0.63, 0.01, 0.00, 0.05, 0.08],
    'F1 Score': [0.50, 0.00, 0.00, 0.00, 0.28, 0.67, 0.01, 0.00, 0.10, 0.12],
    'AP': [0.47, 0.00, 0.00, 0.00, 0.15, 0.63, 0.01, 0.00, 0.04, 0.02]
}

# Converter dados para matriz
metrics = ['Precision', 'Recall', 'F1 Score', 'AP']
matrix = np.array([data['Precision'], data['Recall'], data['F1 Score'], data['AP']])

# Criar heatmap
plt.figure(figsize=(10, 6))
ax = sns.heatmap(matrix.T, annot=True, cmap='viridis', xticklabels=metrics, yticklabels=data['Labels'], cbar=True)

# Adicionar título e rótulos
ax.set_title('Heatmap of Metrics for Pavement Defect Detection')
ax.set_xlabel('Metrics')
ax.set_ylabel('Defect Type')

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
