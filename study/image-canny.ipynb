```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread('roads/IMG_20240218_095141.jpg', 0)  # Carregar a imagem em escala de cinza

# Aplicar o algoritmo de Canny para detectar bordas
bordas = cv2.Canny(imagem, 100, 200)  # 100 e 200 são valores de limiares mínimo e máximo

# Plotar a imagem original
plt.subplot(1, 2, 1)
plt.imshow(imagem, cmap='gray')
plt.title('Imagem Original')
plt.axis('off')

# Plotar as bordas detectadas
plt.subplot(1, 2, 2)
plt.imshow(bordas, cmap='gray')
plt.title('Bordas Detectadas')
plt.axis('off')

plt.show()

```
