```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Função para desenhar a imagem com caixas delimitadoras
def draw_image(objects_to_detect, true_positives, false_positives):
    # Criar uma nova figura
    fig, ax = plt.subplots(1, figsize=(10, 6))
    
    # Desenhar as caixas dos objetos a serem detectados (azul)
    for bbox in objects_to_detect:
        rect = patches.Rectangle((bbox[0]+1, bbox[1]+1), bbox[2], bbox[3], linewidth=2, edgecolor='b', facecolor='none', linestyle='-', label='Objeto a Detectar')
        ax.add_patch(rect)
    
    # Desenhar uma caixa para cada verdadeiro positivo (verde)
    for bbox in true_positives:
        rect = patches.Rectangle((bbox[0]+2, bbox[1]+2), bbox[2], bbox[3], linewidth=2, edgecolor='g', facecolor='none', label='True Positive')
        ax.add_patch(rect)
    
    # Desenhar uma caixa para cada falso positivo (vermelho)
    for bbox in false_positives:
        rect = patches.Rectangle((bbox[0]+2, bbox[1]+15), bbox[2]+5, bbox[3], linewidth=2, edgecolor='r', facecolor='none', linestyle='--', label='False Positive')
        ax.add_patch(rect)
    
    # Ajustar limites e rótulos
    ax.set_xlim(0, 120)
    ax.set_ylim(0, 120)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Detecção de Objetos')
    ax.legend()
    
    # Mostrar a imagem
    plt.gca().invert_yaxis()  # Inverter o eixo Y para coincidir com a coordenada padrão de imagens
    plt.grid(True)
    plt.show()

# Coordenadas das caixas delimitadoras dos objetos a serem detectados (azul)
objects_to_detect = [(20, 20, 20, 20), (60, 40, 25, 25), (90, 10, 20, 20)]  # (x, y, width, height)

# Coordenadas dos verdadeiros positivos (verde), sobrepondo os objetos azuis
true_positives = [(20, 20, 20, 20), (60, 40, 25, 25), (85, 5, 30, 30)]  # (x, y, width, height)

# Coordenadas dos falsos positivos (vermelho), fora dos objetos azuis e verdes
false_positives = [(40, 30, 10, 10), (80, 60, 10, 10)]  # (x, y, width, height)

# Desenhar a imagem
draw_image(objects_to_detect, true_positives, false_positives)
```
