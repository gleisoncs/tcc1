```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import torch

# Função para plotar a imagem com as caixas delimitadoras
def plot_boxes(image_path, boxes):
    # Abre a imagem
    image = Image.open(image_path)
    
    # Plota a imagem
    fig, ax = plt.subplots(1)
    ax.imshow(image)
    
    # Itera sobre as caixas delimitadoras
    for box in boxes:
        x, y, w, h = box
        # Cria um retângulo amarelo com as coordenadas da caixa
        rect = patches.Rectangle((x, y), w, h, linewidth=2, edgecolor='black', facecolor='none')
        # Adiciona o retângulo à plotagem
        ax.add_patch(rect)
    
    # Mostra a plotagem
    plt.show()


# O tensor com as caixas delimitadoras
#tensor = torch.tensor([[ 0.0000,  0.0000,  1.0000,  1.5000]])
tensor = torch.tensor([[0.0, 0.0, 1.0, 1.5],
      [0.0, 0.0, 1.5, 1.5],
      [0.0, 0.0, 2.0, 1.5],
      [0.0, 0.0, 1.5, 2.5],
      [0.0, 0.0, 2.5, 2.5],
      [0.0, 0.0, 3.5, 2.5],
      [0.0, 0.0, 2.0, 3.5],
      [0.0, 0.0, 3.5, 3.5],
      [0.0, 0.0, 5.0, 3.5]])

# Caminho da imagem
image_path = 'random/random11.png'

# Plotar as caixas delimitadoras na imagem
plot_boxes(image_path, tensor.numpy())


tensor = torch.tensor([[0.0, 0.0, 1.0, 1.5],
      [0.0, 0.0, 1.5, 1.5],
      [0.0, 0.0, 2.0, 1.5],
      [0.0, 0.0, 1.5, 2.5],
      [0.0, 0.0, 2.5, 2.5],
      [0.0, 0.0, 3.5, 2.5],
      [0.0, 0.0, 2.0, 3.5],
      [0.0, 0.0, 3.5, 3.5],
      [0.0, 0.0, 5.0, 3.5],
      [0.0, 0.5, 1.0, 2.5],
      [0.0, 0.5, 1.5, 2.5],
      [0.0, 0.5, 2.0, 2.5],
      [0.0, 0.0, 1.5, 3.5],
      [0.0, 0.0, 2.5, 3.5],
      [0.0, 0.0, 3.5, 3.5],
      [0.0, 0.0, 2.0, 4.5],
      [0.0, 0.0, 3.5, 4.5],
      [0.0, 0.0, 5.0, 4.5]])

plot_boxes(image_path, tensor.numpy())

tensor = torch.tensor([[0.0, 0.0, 1.0, 1.5],
      [0.0, 0.0, 1.5, 1.5],
      [0.0, 0.0, 2.0, 1.5],
      [0.0, 0.0, 1.5, 2.5],
      [0.0, 0.0, 2.5, 2.5],
      [0.0, 0.0, 3.5, 2.5],
      [0.0, 0.0, 2.0, 3.5],
      [0.0, 0.0, 3.5, 3.5],
      [0.0, 0.0, 5.0, 3.5],
      [0.0, 0.5, 1.0, 2.5],
      [0.0, 0.5, 1.5, 2.5],
      [0.0, 0.5, 2.0, 2.5],
      [0.0, 0.0, 1.5, 3.5],
      [0.0, 0.0, 2.5, 3.5],
      [0.0, 0.0, 3.5, 3.5],
      [0.0, 0.0, 2.0, 4.5],
      [0.0, 0.0, 3.5, 4.5],
      [0.0, 0.0, 5.0, 4.5],
      [0.0, 1.5, 1.0, 3.5],
      [0.0, 1.5, 1.5, 3.5],
      [0.0, 1.5, 2.0, 3.5],
      [0.0, 0.5, 1.5, 4.5],
      [0.0, 0.5, 2.5, 4.5],
      [0.0, 0.5, 3.5, 4.5],
      [0.0, 0.0, 2.0, 5.5],
      [0.0, 0.0, 3.5, 5.5],
      [0.0, 0.0, 5.0, 5.5]])

plot_boxes(image_path, tensor.numpy())

```
