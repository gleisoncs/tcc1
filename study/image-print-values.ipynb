```python
from PIL import Image
import numpy as np

# Carregar a imagem
image_path = "random/elephant.jpeg"
image = Image.open(image_path)

# Converter a imagem para escala de cinza (se necess√°rio) e redimensionar para 28x28 (para corresponder ao formato MNIST)
image = image.convert('L').resize((28, 28))

# Converter a imagem para uma matriz numpy
image_array = np.array(image)

# Imprimir cada linha da imagem em uma linha separada
for row in image_array:
    print(' '.join([f'{pixel:3}' for pixel in row]))
```

```python
image2 = image.convert('RGB').resize((28, 28))
image_array = np.array(image2)

# Imprimir os valores dos tr√™s canais para cada pixel
for row in image_array:
    for pixel in row:
        print(f'({pixel[0]:3}, {pixel[1]:3}, {pixel[2]:3})', end=' ')
    print()
```
