```python
import numpy as np
import matplotlib.pyplot as plt

# Função de perda simples: função quadrática
def loss_function(x):
    return x**2

# Gradiente da função de perda: derivada da função quadrática
def gradient(x):
    return 2 * x

# Algoritmo de Gradiente Descendente com Momentum
def gradient_descent_momentum(starting_point, learning_rate, momentum, num_iterations):
    x = starting_point  # Ponto inicial
    v = 0  # Inicialização da velocidade (momentum)
    
    trajectory = []  # Armazena a trajetória do ponto para visualização

    for i in range(num_iterations):
        grad = gradient(x)  # Calcula o gradiente da função de perda no ponto atual
        v = momentum * v - learning_rate * grad  # Atualiza a "velocidade" com o momentum
        x = x + v  # Atualiza a posição com base na velocidade

        # Armazena a trajetória
        trajectory.append(x)
        
        # Observação: Exibir a cada 10 iterações
        if i % 10 == 0:
            print(f"Iteração {i}: x = {x}, grad = {grad}, v = {v}")

    return x, trajectory
```

```python
# Parâmetros
starting_point = 20  # Começamos longe do mínimo (que é em x = 0)
learning_rate = 0.1  # Taxa de aprendizado
momentum = 0.82  # Coeficiente de momentum
num_iterations = 50  # Número de iterações

# Executa o gradiente descendente com momentum
final_x, trajectory = gradient_descent_momentum(starting_point, learning_rate, momentum, num_iterations)

# Exibe o resultado final
print(f"\nPonto final após {num_iterations} iterações: x = {final_x}")

# Visualização da trajetória
plt.plot(range(num_iterations), trajectory, marker='o')
plt.xlabel('Iterações')
plt.ylabel('Valor de x')
plt.title('Trajetória do Gradiente Descendente com Momentum')
plt.grid()
plt.show()
```
