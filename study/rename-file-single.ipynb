```python
import os

def rename_png_to_xml(directory):
    # Percorre todos os arquivos no diretrio
    for filename in os.listdir(directory):
        if filename.endswith('.png'):
            # Monta o caminho completo do arquivo original
            old_file_path = os.path.join(directory, filename)
            
            # Gera o novo nome de arquivo com extenso .xml
            new_filename = os.path.splitext(filename)[0] + '.xml'
            
            # Monta o caminho completo do novo arquivo
            new_file_path = os.path.join(directory, new_filename)
            
            # Renomeia o arquivo
            os.rename(old_file_path, new_file_path)
            print(f"Arquivo {filename} renomeado para {new_filename}")

# Diretrio contendo os arquivos PNG
directory_path = 'C:/Users/caetano/Downloads/job14/Annotations/masks-only/'

# Chama a funo para renomear os arquivos
rename_png_to_xml(directory_path)
```
