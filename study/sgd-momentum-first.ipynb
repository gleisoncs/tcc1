```python
# Função simples de exemplo para otimização (f(x) = (x - 2)^2)
def func(x):
    return (x - 2) ** 2

# Gradiente da função (derivada de f(x))
def grad(x):
    return 2 * (x - 2)

# SGD com Momentum
def sgd_momentum(learning_rate, momentum, num_iterations):
    x = 0  # Começamos em 0 (posição inicial)
    velocity = 0  # Inicializamos a velocidade

    positions = []  # Para guardar o histórico de posições

    for i in range(num_iterations):
        print('----')
        print('x', x)
        gradient = np.round(grad(x), 2)  # Calcula o gradiente (quão íngreme é a "colina")
        print('gradient', gradient)
        # Atualiza a velocidade com base no gradiente e no momentum
        velocity = np.round(momentum * velocity - learning_rate * gradient, 2)
        print('velocity', velocity)
        # Atualiza a posição (ou seja, ajusta o valor de x)
        x = np.round(x + velocity, 2)
        print('x', x)
        positions.append(x)  # Guarda a posição

    return positions
```

```python
# Parâmetros do exemplo
learning_rate = 0.1  # Taxa de aprendizado (quão rápido ajustamos)
momentum = 0.9  # Quanto de velocidade acumulada usamos
num_iterations = 5  # Número de passos

# Rodamos o SGD com momentum
positions = sgd_momentum(learning_rate, momentum, num_iterations)
```

```python
#PRIMEIRO

import numpy as np
import matplotlib.pyplot as plt

# Função simples de exemplo para otimização (f(x) = (x - 2)^2)
def func(x):
    return (x - 2) ** 2

# Gradiente da função (derivada de f(x))
def grad(x):
    return 2 * (x - 2)

# SGD com Momentum
def sgd_momentum(learning_rate, momentum, num_iterations):
    x = 0  # Começamos em 0 (posição inicial)
    velocity = 0  # Inicializamos a velocidade

    positions = []  # Para guardar o histórico de posições

    for i in range(num_iterations):
        gradient = grad(x)  # Calcula o gradiente (quão íngreme é a "colina")

        # Atualiza a velocidade com base no gradiente e no momentum
        velocity = momentum * velocity - learning_rate * gradient

        # Atualiza a posição (ou seja, ajusta o valor de x)
        x = x + velocity

        positions.append(x)  # Guarda a posição

    return positions

# Parâmetros do exemplo
learning_rate = 0.1  # Taxa de aprendizado (quão rápido ajustamos)
momentum = 0.9  # Quanto de velocidade acumulada usamos
num_iterations = 50  # Número de passos

# Rodamos o SGD com momentum
positions = sgd_momentum(learning_rate, momentum, num_iterations)

# Plotando o gráfico para ver como a otimização se comporta
x_vals = np.linspace(-2, 6, 100)
y_vals = func(x_vals)

plt.plot(x_vals, y_vals, label="f(x) = (x - 2)^2")
plt.plot(positions, func(np.array(positions)), 'o-', label="Caminho do SGD com Momentum", color="red")
plt.xlabel("Posição x")
plt.ylabel("f(x)")
plt.legend()
plt.title("Exemplo de SGD com Momentum")
plt.show()
```
