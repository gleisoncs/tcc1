```python
import matplotlib.pyplot as plt
import numpy as np

# Definindo as coordenadas da primeira caixa delimitadora (x1, y1, x2, y2)
box1 = np.array([100, 100, 300, 300])

# Definindo as coordenadas da segunda caixa delimitadora (x1, y1, x2, y2)
box2 = np.array([200, 200, 400, 400])

# Calculando os pontos de interseção
inter_rect_x1 = max(box1[0], box2[0])
inter_rect_y1 = max(box1[1], box2[1])
inter_rect_x2 = min(box1[2], box2[2])
inter_rect_y2 = min(box1[3], box2[3])

# Calculando a largura e altura da interseção
inter_width = max(inter_rect_x2 - inter_rect_x1 + 1, 0)
inter_height = max(inter_rect_y2 - inter_rect_y1 + 1, 0)

# Área de interseção
inter_area = inter_width * inter_height

# Criando a figura e os eixos
fig, ax = plt.subplots()

# Desenhando as caixas delimitadoras
rect1 = plt.Rectangle((box1[0], box1[1]), box1[2] - box1[0], box1[3] - box1[1], linewidth=1, edgecolor='r', facecolor='none')
rect2 = plt.Rectangle((box2[0], box2[1]), box2[2] - box2[0], box2[3] - box2[1], linewidth=1, edgecolor='b', facecolor='none')

# Adicionando as caixas delimitadoras aos eixos
ax.add_patch(rect1)
ax.add_patch(rect2)

# Plotando os pontos de interseção se houver interseção
if inter_width > 0 and inter_height > 0:
    inter_rect = plt.Rectangle((inter_rect_x1, inter_rect_y1), inter_width, inter_height, linewidth=1, edgecolor='g', facecolor='none')
    ax.add_patch(inter_rect)
    ax.text(inter_rect_x1 + inter_width / 2, inter_rect_y1 + inter_height / 2, f'Área de Interseção: {inter_area} pixels quadrados', ha='center', va='center', color='g')

# Configurando os limites dos eixos
ax.set_xlim(0, 500)
ax.set_ylim(0, 500)

# Configurando o título e rótulos dos eixos
plt.title('Exemplo de Interseção entre Caixas Delimitadoras')
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')

ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')
    
# Exibindo o gráfico
plt.show()
```
