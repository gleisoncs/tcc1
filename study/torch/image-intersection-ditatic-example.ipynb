```python
import matplotlib.pyplot as plt
import torch
```

```python
# Definindo as coordenadas dos quadrados (x1, y1, x2, y2)
box1 = torch.tensor([2, 2, 4, 4])
box2 = torch.tensor([3, 3, 5, 5])

# Calculando os pontos de interseção
inter_rect_x1 = torch.max(box1[0], box2[0])
inter_rect_y1 = torch.max(box1[1], box2[1])
inter_rect_x2 = torch.min(box1[2], box2[2])
inter_rect_y2 = torch.min(box1[3], box2[3])

# Calculando a largura e altura da interseção
inter_width = torch.clamp(inter_rect_x2 - inter_rect_x1 + 1, min=0)
inter_height = torch.clamp(inter_rect_y2 - inter_rect_y1 + 1, min=0)

# Área de interseção
inter_area = inter_width * inter_height

# Criando a figura e os eixos
fig, ax = plt.subplots()

# Desenhando os quadrados
rect1 = plt.Rectangle((box1[0], box1[1]), box1[2] - box1[0], box1[3] - box1[1], linewidth=1, edgecolor='r', facecolor='none')
rect2 = plt.Rectangle((box2[0], box2[1]), box2[2] - box2[0], box2[3] - box2[1], linewidth=1, edgecolor='b', facecolor='none')

# Adicionando os quadrados aos eixos
ax.add_patch(rect1)
ax.add_patch(rect2)

# Plotando os pontos de interseção se houver interseção
if inter_width > 0 and inter_height > 0:
    inter_rect = plt.Rectangle((inter_rect_x1, inter_rect_y1), inter_width, inter_height, linewidth=1, edgecolor='g', facecolor='none')
    ax.add_patch(inter_rect)
    ax.text(inter_rect_x1 + inter_width / 2, inter_rect_y1 + inter_height / 2, f'Intersection Area: {inter_area.item()} sq pixels', ha='center', va='center', color='g')

# Configurando os limites dos eixos (invertendo os eixos)
ax.set_xlim(0, 7)
ax.set_ylim(7, 0)

# Configurando o título e rótulos dos eixos
ax.set_title('Exemplo de Interseção entre Quadrados')
ax.set_xlabel('Coordenada X')
ax.set_ylabel('Coordenada Y')

# Configurando os eixos para ficarem em cima
ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')

# Exibindo o gráfico
plt.show()
```
