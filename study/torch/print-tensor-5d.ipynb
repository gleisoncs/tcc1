```python
import torch

# Create a sample tensor with 5 dimensions
# Replace the shape with your actual tensor shape
#tensor_5d = torch.randn(2, 3, 4, 5, 6)
tensor_5d = torch.randint(low=0, high=10, size=(2, 3, 4, 5, 6))


# Print a portion of the tensor
print(tensor_5d[:, :, :, :, :])  # Slice along the 3rd dimension and print the rest

# OR

# If you want to print the whole tensor, you might need to iterate over each dimension
# and print slices of the tensor.

# Iterate over each element of the tensor
for i in range(tensor_5d.size(0)):
    for j in range(tensor_5d.size(1)):
        for k in range(tensor_5d.size(2)):
            for l in range(tensor_5d.size(3)):
                for m in range(tensor_5d.size(4)):
                    print("Element at index ({}, {}, {}, {}, {}): {}".format(i, j, k, l, m, tensor_5d[i, j, k, l, m]))
```
